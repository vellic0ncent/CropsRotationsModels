```python
import pandas as pd
```

```python
df2018 = pd.read_csv('./data/test 2019/test_2018.csv')
df2018
```

```python
df2017 = pd.read_csv('./data/test 2019/test_2017.csv')
df2017
```

```python
df2016 = pd.read_csv('./data/test 2019/test_2016.csv')
df2016.rename(columns={'CODE_CULTU':'CODE_CULTU_2016', 'CODE_GROUP': 'CODE_GROUP_2016'}, inplace=True)
df2016
```

```python
df = df.merge(df2016, on='centroid')
del df2016
df
```

```python
df2015 = pd.read_csv('./data/test 2019/test_2015.csv')
df2015.rename(columns={'CODE_CULTU':'CODE_CULTU_2015', 'CODE_GROUP': 'CODE_GROUP_2015'}, inplace=True)
df2015
```

```python
df = df.merge(df2015, on='centroid')
del df2015
df
```

```python
from shapely.geometry import Point, Polygon
import geopandas as gpd
import shapely.wkt
```

```python
cdf = df[['centroid']]
cdf['wgs84'] = gpd.GeoSeries.from_wkt(cdf['centroid'], crs='epsg:2154').to_crs('epsg:4326')
```

```python
cdf
```

```python
fr_phz = gpd.read_file('./data/geojson/FR_PHZ.geojson')
```

```python
cdf = gpd.GeoDataFrame(cdf, geometry='wgs84', crs='EPSG:4326')
cdf = gpd.sjoin(cdf, fr_phz, op='within')
cdf
```

```python
cdf.drop(['pm_icon', 'index_right'], axis=1, inplace=True)
```

```python
kg = gpd.read_file('./data/geojson/eur_kg.geojson')
```

```python
cdf = gpd.sjoin(cdf, kg, op='within')
cdf
```

```python
cdf.drop(['index_right', 'Shape_Leng', 'Shape_Area', 'pnm'], axis=1, inplace=True)
```

```python
fr_ff = gpd.read_file('./data/geojson/FR_ff.geojson')
```

```python
cdf = gpd.sjoin(cdf, fr_ff, op='within')
cdf
```

```python
cdf.drop(['index_right', 'pm_icon'], axis=1, inplace=True)
cdf
```

```python
cdf.drop('wgs84', axis=1, inplace=True)
```
