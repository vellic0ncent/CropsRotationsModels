```python
import pandas as pd
```

```python
df2015 = pd.read_csv('train/train_2015.csv')
df2015.columns.values[0] = '2015'

df2016 = pd.read_csv('train/train_2016.csv')
df2016.columns.values[0] = '2016'

df2017 = pd.read_csv('train/train_2017.csv')
df2017.columns.values[0] = '2017'

df2018 = pd.read_csv('train/train_2018.csv')
df2018.columns.values[0] = '2018'

df2019 = pd.read_csv('train/train_2019.csv')
df2019.columns.values[0] = '2019'
```

```python
df15_16 = pd.merge(df2015, df2016, how='inner', left_on='centroid',
right_on = 'centroid')

df15_16_17 = pd.merge(df15_16, df2017, how='inner', left_on='centroid',
right_on = 'centroid')

df15_16_17_18 = pd.merge(df15_16_17, df2018, how='inner', left_on='centroid',
right_on = 'centroid')

df = pd.merge(df15_16_17_18, df2019, how='inner', left_on='centroid',
right_on = 'centroid')
df.columns
```

```python
df =df.drop(['CODE_GROUP_x', 'CODE_GROUP_y','CODE_GROUP_x', 'CODE_GROUP_y','CODE_GROUP'], axis = 1)
```

```python
columnsTitles=["centroid","2015","2016","2017","2018","2019"]

df=df.reindex(columns=columnsTitles)

test_df = df.drop(['centroid'], axis = 1)
```

```python
# Длина ротации
def rotationLength(row):
  length = row.drop_duplicates().shape[0]
  return length

# Число ротаций
def rotationCount(row):
  count = row.value_counts().min()
  return count
```

```python
rotation_length = test_df.apply(rotationLength, axis=1)

rotation_count = test_df.apply(rotationCount, axis=1)
```

```python
df['RotationLength'] = rotation_length
df['RotationCount'] = rotation_count
df
```

```python
df.to_csv('features_RotationLength_RotationCount_train.csv', index = False) # сохраняем CSV файл
```
