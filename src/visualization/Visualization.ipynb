```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.cm as cmx

from pyproj import Proj, transform, Transformer
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
data = pd.read_csv('/content/drive/MyDrive/Хакатон/dataset2015_2019.csv')
```

```python
data.head(2)
```

```python
#Получение географических координат
transformer = Transformer.from_crs(2154, 4326, always_xy=True)
x, y = transformer.transform(data["x"].values, data["y"].values)
data['X_coord'] = x
data['Y_coord'] = y
```

```python
#Визуализация по кодам групп
plt.rcParams['figure.figsize'] = 15, 15
plt.scatter(data['X_coord'], data['Y_coord'], c=data['CODE_GROUP_2019'], label=data['CODE_GROUP_2019'], cmap='viridis', alpha=0.1)
plt.show()
```
